### 1. B

#### Query : 
`Return the most number of catches taken by a player in IPL history`

```python
catches = ipl_ball_by_ball[ipl_ball_by_ball["dismissal_kind"] == 'caught'].groupby('fielder').count()
c_and_b = ipl_ball_by_ball[ipl_ball_by_ball["dismissal_kind"] == 'caught and bowled'].groupby('bowler').count()
type(list(c_and_b.index))
result_df1 = pd.DataFrame(
    data={'Name':list(c_and_b.index), 'Count':list(c_and_b['player_dismissed'])}, 
    index=[_ for _ in range(len(c_and_b))])
result_df2 = pd.DataFrame(
    data={'Name':list(catches.index), 'Count':list(catches['player_dismissed'])}, 
    index=[_ for _ in range(len(catches))])
result = pd.merge(result_df1, result_df2, how="outer")
result.sort_values('Count', ascending=False)[:3]
```

```python
catches_max = ipl_ball_by_ball[ipl_ball_by_ball["dismissal_kind"].isin(['caught','caught and bowled'])].groupby('fielder').count()
catches_max = catches_max.sort_values('dismissal_kind', ascending=False)[:1]['is_wicket']
print(f"Player {catches_max.index[0]} took a total of {catches_max[0]} catches, highest in the IPL history ")
```

#### Query:
`Write a query to return a report for highest wicket taker in matches which were affected by Duckworth-Lewis’s method (D/L method).`

```python
temp2 = ipl_matches[ipl_matches['method'] == 'D/L']
matches_affected_by_dl = list(temp2['match_id'])
temp2 = ipl_ball_by_ball[ipl_ball_by_ball['match_id'].isin(matches_affected_by_dl)]
temp2 = temp2[temp2["is_wicket"]==1].groupby('bowler').count()
temp2.sort_values('is_wicket', ascending=False)['is_wicket']
```
